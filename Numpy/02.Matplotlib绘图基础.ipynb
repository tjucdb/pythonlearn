# 简介
Matplotlib是一个python绘图十分简单、有用的库，和matlab有点类似。

在ipython中可以使用魔术命令import相关库

`%matplotlib inline` 这个命令会让图像直接显示在jupyter notebook中

`%matplotlib` 在新窗口弹出图像

```python
%pylab
%matplotlib inline
```

## 二维图

```python
x = linspace(0, 2*pi, 100)
y = sin(x)
plot(x, y)  
```

### 改变颜色、线型、线宽
与matlab使用方法基本一致。

```python
plot(x,sin(x),'r-')
```

或者可以用下面的方法指定颜色，线宽，线型：

```python
plot(x, sin(x), color='green', linewidth=3, linestyle='-')
```

### 改变x,y轴的显示

```python
x = linspace(0, 2*pi, 100)
y = sin(x)
xlim(x.min() * 1.1, x.max() * 1.1)
ylim(y.min() * 1.1, y.max() * 1.1)
plot(x, y)  
```

```python
xticks([0, pi/2, pi, 3*pi/2,2*pi])
yticks([-1, 0, +1])
plot(x,y)
```

```python
xticks([0, pi/2, pi, 3*pi/2,2*pi],
       [r'$0$', r'$+\pi/2$', r'$+\pi$', r'$+3\pi/2$', r'$+2\pi$'])
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])
plot(x,y)
```

### 多条图像

```python
plot(x,sin(x),
     x,cos(x))
```

用figure命令产生新的图像：

```python
t = linspace(0, 2*pi, 100)
x = sin(t)
y = cos(t)
figure()
plot(t,x)
figure()
plot(t,y)
```

或者使用 subplot 在一幅图中画多幅子图：
subplot(row, column, index)

```python
subplot(1, 2, 1)
plot(t,x)
subplot(1, 2, 2)
plot(t,y)
```

## 添加标签、网格、标题

```python
plot(t,x, label='sin(x)')
plot(t,y, label='cos(x)')
xlabel('radians')
ylabel('number')
legend(loc='upper right') #loc标签位置
grid() #添加网格
title('Test Matpoltlib') #添加标题
```
